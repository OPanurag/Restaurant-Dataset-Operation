```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Example DataFrame with longitude and latitude columns
df = pd.read_csv("./L1T2_Dataset.csv")
```

```python
# Create a GeoDataFrame from the DataFrame
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df['Longitude'], df['Latitude']))
```

```python
# Load the world map data from GeoPandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
# Plot the world map
ax = world.plot(figsize=(30, 50), color='lightgrey')

# Plot the restaurant locations on the map
gdf.plot(ax=ax, color='red', marker='o', markersize=10)

# # Add labels for each restaurant
# for x, y, label in zip(df['Longitude'], df['Latitude'], df['Restaurant Name']):
#     ax.text(x, y, label, fontsize=8)

# Show the map
plt.show()
```
