```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./L1T2_Dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
price_range_dict = {}

for price_range in df['Price range'].unique():
    restaurants_in_range = list(df.loc[df['Price range'] == price_range, 'Restaurant Name'])
    price_range_dict[price_range] = restaurants_in_range
```

```python
sorted_keys = sorted(price_range_dict.keys())
sorted_dict = {key: price_range_dict[key] for key in sorted_keys}
sorted_dict
```

```python
average_rating_by_price = df.groupby('Price range')['Rating text'].mean()
average_rating_dict = average_rating_by_price.to_dict()
average_rating_dict
```

```python
max_rating_indices = df.groupby('Price range')['Rating text'].idxmax()

# Create a new DataFrame with the rows containing the highest rating color for each group
max_rating_color_df = df.loc[max_rating_indices, ['Price range', 'Rating color']]
```
